```python
# Import libraries 
import numpy as np
import pandas as pd
```

```python
# Set path to data file
path = "transactions.txt"
```

```python
# Load data into dataframe
df = pd.read_json(path, lines=True)
# Replace empty data with NaN
df.replace('', np.nan, inplace=True)
df.head()
```

```python
# Look at structure of data.
print(df.info())
```

```python
# Count of null values for each column
print(df.isnull().sum())
```

```python
# Drop empty columns with no data.
df.drop(['echoBuffer','merchantCity', 'merchantState', 'merchantZip', 'posOnPremises', 'recurringAuthInd'], axis=1, inplace=True)
```

```python
# ount of null values for each column
print(df.isnull().sum())
```

```python
# Reset index of dataframe.
df = df.reset_index(drop=True)
df
```

```python
# Pickle cleaned data to file for later use
pd.to_pickle(df, 'data/cleaned_data', protocol=4)
```
